# Symmetric function exploration
# elementary, Schur, monomial, homogenous, powersum)

For definitions see:
https://www2.math.upenn.edu/~peal/polynomials/standardSymmetricFunctions.htm#elementaryE

```
User chooses nmax and list of basis_types from ['m','e','s','p','h'] (default is basis_types='m') and myalphabet (default is my_alphabet=['x1','x2','x3','x4','x5'])
```

````python
def sym_explore(basis_types = ['m'],nmax=4,my_alphabet=['x1','x2','x3','x4']):
    kmax=len(my_alphabet)
    for type in basis_types:
        if type=='m':
            b = SymmetricFunctions(QQ).monomial() # b for basis
        elif type=='e':
            b = SymmetricFunctions(QQ).elementary()
        elif type=='s':
            b = SymmetricFunctions(QQ).schur()
        elif type=='p':
            b = SymmetricFunctions(QQ).powersum()
        elif type=='h':
            b = SymmetricFunctions(QQ).homogeneous()
        else:
            raise ValueError("The argument basis_types includes",type,"which is not implemented.")
    for k in range(1,kmax+1):
        for n in range(1,k+1):
            for Lambda in Partitions(n):
                #show(b(Lambda),tuple(my_alphabet[:k]),'=',b(Lambda).expand(k, alphabet=my_alphabet[:k]),'\n')
                show(b(Lambda),'=',b(Lambda).expand(k, alphabet=my_alphabet[:k]),'\n')
````

````python
sym_explore()
````

````python
sym_explore(['e'])
````

````python
sym_explore(['s'])
````

````python
sym_explore(['p'])
````

````python
sym_explore(['h'])
````
